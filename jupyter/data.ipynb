```python
continuous_vars = ['bp_systolic', 'bp_diastolic', 'blood_urea', 
                   'serum_creatinine', 'blood_glucose_random',
                   'bioavailability_pct', 'volume_of_distribution',
                   'kidney_cell_viability_pct', 'toxicity_score_composite',
                   'pk_toxic_interaction_score']
df[continuous_vars].describe()
```

```python
def validate_cleaned_data(df):
    """Validate the cleaned dataset"""
    
    print("=== DATA VALIDATION REPORT ===")
    
    # 1. Check for remaining negatives in biological measurements
    biological_cols = ['blood_urea', 'serum_creatinine', 'blood_glucose_random']
    for col in biological_cols:
        if col in df.columns:
            negatives = (df[col] < 0).sum()
            print(f"{col}: {negatives} negative values remaining")
    
    # 2. Check percentage bounds
    percentage_cols = ['kidney_cell_viability_pct', 'protein_binding_pct', 'bioavailability_pct']
    for col in percentage_cols:
        if col in df.columns:
            below_zero = (df[col] < 0).sum()
            above_100 = (df[col] > 100).sum()
            print(f"{col}: {below_zero} below 0, {above_100} above 100")
    
    # 3. Check missing values
    missing = df.isnull().sum().sum()
    print(f"Total missing values: {missing}")
    
    # 4. Check data types
    print("\nData types:")
    print(df.dtypes.value_counts())
    
    # 5. Basic statistics
    print("\nBasic statistics for key columns:")
    key_cols = ['patient_age', 'bp_systolic', 'bp_diastolic', 'blood_urea', 'serum_creatinine']
    print(df[key_cols].describe())
    
    return True

# Run validation
validate_cleaned_data(df)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Histograms for continuous variables
fig, axes = plt.subplots(4, 3, figsize=(15, 12))
axes = axes.ravel()
for idx, col in enumerate(continuous_vars[:12]):
    axes[idx].hist(df[col], bins=30, edgecolor='black', alpha=0.7)
    axes[idx].set_title(f'Distribution of {col}')
    axes[idx].set_xlabel(col)
    axes[idx].set_ylabel('Frequency')
plt.tight_layout()
plt.show()

# 2. Box plots by CKD risk label
fig, axes = plt.subplots(4, 3, figsize=(15, 12))
axes = axes.ravel()
for idx, col in enumerate(continuous_vars[:12]):
    sns.boxplot(x='ckd_risk_label', y=col, data=df, ax=axes[idx])
    axes[idx].set_title(f'{col} by CKD Risk')
plt.tight_layout()
plt.show()
```

```python
import numpy as np 
# Select all numeric variables
numeric_df = df.select_dtypes(include=[np.number])

# Heatmap of correlations
plt.figure(figsize=(16, 12))
correlation_matrix = numeric_df.corr()
sns.heatmap(correlation_matrix, annot=False, cmap='coolwarm', 
            center=0, linewidths=0.5)
plt.title('Correlation Matrix Heatmap')
plt.show()

# Top correlations with CKD risk
ckd_correlations = correlation_matrix['ckd_risk_label'].sort_values(ascending=False)
print("Top correlations with CKD Risk Label:")
print(ckd_correlations.head(10))
print("\nBottom correlations with CKD Risk Label:")
print(ckd_correlations.tail(10))
```
